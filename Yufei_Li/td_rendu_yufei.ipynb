```python
# Exercice 1
import requests

# Ask for first name
first_name = input("Please enter your first name: ")

# GET request to Agify API
url = f"https://api.agify.io/?name={first_name}"
response = requests.get(url)

# Check
if response.status_code == 200:
    # Parse the response data
    data = response.json()

    # Extract the age 
    age = data["age"]
    count = data["count"]

    # Create a personalized response
    response_data = {
        "name": first_name,
        "age": age,
        "count": count
    }

    # Print 
    print(response_data)
else:
    print("Error: Failed to retrieve data from the Agify API.")
```

```python
# Exercice 2
import requests

# Ask first name
first_name = input("Please enter your first name: ")

#  GET request to genderize API
url = f"https://api.genderize.io/?name={first_name}"
response = requests.get(url)

# Check 
if response.status_code == 200:
    
    data = response.json()

    # Extract the gender 
    gender = data["gender"]
    probability = data["probability"]

    # Create a personalized message
    if gender is not None:
        message = f"Hi {first_name}, according to the Genderize API, the estimated gender for someone named {first_name} is {gender} with a probability of {probability}."
    else:
        message = f"Hi {first_name}, the Genderize API couldn't determine the gender for someone named {first_name}."

    # Print 
    print(message)
else:
    print("Error: Failed to retrieve data from the Genderize API.")
```

```python
# Exercice 3(1)
import requests

# Ask first name
first_name = input("Please enter your first name: ")

# GET request to the Nationalize API
url = f"https://api.nationalize.io/?name={first_name}"
response = requests.get(url)

# Check 
if response.status_code == 200:
    
    data = response.json()

    # Extract 
    country_predictions = data["country"]

    # Create a personalized message
    if len(country_predictions) > 0:
        countries = [prediction["country_id"] for prediction in country_predictions]
        probabilities = [prediction["probability"] for prediction in country_predictions]
        message = f"Hi {first_name}, according to the Nationalize API, your name {first_name} is most likely associated with the following countries: {', '.join(countries)} with corresponding probabilities: {', '.join(map(str, probabilities))}."
    else:
        message = f"Hi {first_name}, the Nationalize API couldn't predict any country associated with the name {first_name}."

    # Print 
    print(message)
else:
    print("Error: Failed to retrieve data from the Nationalize API.")
```

```python
# Exercice 3（2）
import requests
import json

def get_nationality(name):
    url = f"https://api.nationalize.io/?name={name}"
    response = requests.get(url)
    data = response.json()
    return data

name = input("Please enter your first name: ")
nationality_data = get_nationality(name)

if 'country' in nationality_data:
    most_probable_country = max(nationality_data['country'], key=lambda x: x['probability'])
    country_code = most_probable_country['country_id']
    probability = most_probable_country['probability']
    country_name = most_probable_country.get('country_name', '.')
    
    message = f"Name: {name}\nMost probable country: {country_code} ({country_name})\nProbability: {probability:.2f}"
else:
    message = f"Hello, {name}! We couldn't determine your nationality with the provided name."

print(message)
```

```python
# Exercice 7

import requests
import csv

# Create a python class
# Based on what I've done before for the exercise
class ManorStats:
    def __init__(self):
        self.opendomesday_url = "https://opendomesday.org/api/1.0"
        self.counties_endpoint = "/county"
        self.manor_stats_file = "manor_stats.csv"
        self.counties = None
        self.derbyshire = None
        self.places = None
        self.liste_manor_infos = None

        
    # Get countries
    def fetch_counties(self):
        response = requests.get(f"{self.opendomesday_url}{self.counties_endpoint}")
        if response.status_code == 200:
            self.counties = response.json()
        else:
            print("Error fetching counties.")

    # Let's us find Derbyshire
    def find_derbyshire(self):
        if self.counties:
            for county in self.counties:
                if county["name"] == "Derbyshire":
                    self.derbyshire = county
                    break
        else:
            print("Counties data not available.")

    def get_places_in_derbyshire(self):
        if self.derbyshire:
            self.places = [place["id"] for place in self.derbyshire.get("places_in_county")]
        else:
            print("Derbyshire data not available.")
            
            
    # Fetch
    def fetch_manor_data(self):
        if self.places:
            self.liste_manor_infos = []
            for place_id in self.places:
                url = f"{self.opendomesday_url}/manor/{place_id}"
                response = requests.get(url)
                if response.status_code == 200:
                    self.liste_manor_infos.append(response.json())
                else:
                    print(f"Error fetching manor data for place ID: {place_id}")
        else:
            print("Places data not available.")
            
            
    # Save to csv
    def save_manor_stats_to_csv(self):
        if self.liste_manor_infos:
            data_rows = []
            for manor_info in self.liste_manor_infos:
                if manor_info.get("county") == "Derbyshire":
                    manor_name = manor_info.get("name")
                    total_tax = manor_info.get("total_tax")
                    ploughs = manor_info.get("ploughs")
                    data_rows.append([manor_name, total_tax, ploughs])

            with open(self.manor_stats_file, "w", newline="") as file:
                writer = csv.writer(file)
                writer.writerow(["Manor Name", "Total Tax", "Ploughs"])
                writer.writerows(data_rows)

            print(f"Manor stats saved to {self.manor_stats_file}")
        else:
            print("Manor data not available.")

    # Find richest
    def find_richest_manor(self):
        richest_manor = None
        richest_value = 0

        if self.liste_manor_infos:
            for manor_info in self.liste_manor_infos:
                if manor_info.get("county") == "Derbyshire":
                    total_tax = manor_info.get("total_tax")
                    if total_tax > richest_value:
                        richest_manor = manor_info.get("name")
                        richest_value = total_tax

            if richest_manor:
                print(f"The richest manor in Derbyshire is: {richest_manor}")
            else:
                print("No richest manor found.")
        else:
            print("Manor data not available.")
    
    # Total value paid
    def calculate_total_value_paid(self):
        total_value_paid = 0

        if self.liste_manor_infos:
            for manor_info in self.liste_manor_infos:
                if manor_info.get("county") == "Derbyshire":
                    total_tax = manor_info.get("total_tax")
                    total_value_paid += total_tax

            print(f"The total value paid by Derbyshire is: {total_value_paid}")
        else:
            print("Manor data not available.")


# Create instance of ManorStats class
manor_stats = ManorStats()

# Fetch counties data
manor_stats.fetch_counties()

# Find Derbyshire in the counties
manor_stats.find_derbyshire()

# Get the places in Derbyshire
manor_stats.get_places_in_derbyshire()

# Fetch manor data for Derbyshire
manor_stats.fetch_manor_data()

# Save manor to CSV 
manor_stats.save_manor_stats_to_csv()

# Richest manor in Derbyshire
manor_stats.find_richest_manor()

# Total value paid by Derbyshire
manor_stats.calculate_total_value_paid()
```
